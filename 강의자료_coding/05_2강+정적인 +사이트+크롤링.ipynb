```python
# 2. 원하는 데이터 추출
```

```python
from bs4 import BeautifulSoup
```

```python
dom = BeautifulSoup(response.text, 'html.parser')
```

```python
post = dom.select_one("a.post-title")
```

```python
post.text
```

```python
post.attrs.get("href")
```

```python
posts = dom.select("div.post-listing section.post")
```

```python
len(posts)
```

```python
posts[0].select_one("a.post-title")
```

```python
for post in posts:
    post = post.select_one("a.post-title")
    title = post.text
    url = post.attrs.get("href")
    print(title, url)
```

```python
# 네이버 실시간 급상승 검색어 크롤링
import requests
url = "https://www.naver.com/"
```

```python
res = requests.get(url)
```

```python
res
```

```python
from bs4 import BeautifulSoup
```

```python
dom = BeautifulSoup(res.text, 'html.parser')
```

```python
keywords = dom.select("ul.ah_l li.ah_item")
```

```python
len(keywords)
```

```python
keywords[:20]
```

```python
file = open("rank.csv", 'w')
file.write("rank,keyword\n")
for item in keywords[:20]:
    rank = item.select_one("span.ah_r").text
    keyword = item.select_one("span.ah_k").text
    
    file.write("{},{}\n".format(rank, keyword))
    
    print(rank, keyword)
    
file.close()
```
