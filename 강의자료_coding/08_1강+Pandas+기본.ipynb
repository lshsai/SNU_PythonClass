```python
# 전화번호
phone = pd.DataFrame(columns=["Phonenumber"])
```

```python
phone
```

```python
phone.loc[0] = "010-5032-9432"
```

```python
phone.loc[1] = "공일공-오공삼이-9432"
```

```python
phone.loc[2] = "010.5032.구사삼이"
```

```python
phone.loc[3] = "공1공-5032.구사삼2"
```

```python
def get_preprocessed_phone(phonenumber):
    
    _dict = {
        '공': '0',
        '일': '1',
        '이': '2',
        '삼': '3',
        '사': '4',
        '오': '5',
        # ...
        '구': '9',
        " ": "",
        "-": "",
        ".": ""
    }
    
    for key, value in _dict.items():
        phonenumber = phonenumber.replace(key, value)
    return phonenumber
```

```python
phone["Preprocessed Phonenumber"] = phone["Phonenumber"].apply(get_preprocessed_phone)
```
