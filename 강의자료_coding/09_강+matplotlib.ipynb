```python
fig, ax = plt.subplots()
x = np.arange(10)
y = x ** 2

ax.plot(
    x, y,
    linestyle=":",  # solid ( - ), dashed( -- ), dashdot( -. ), dotted( : )
    marker="*",
    color="0.5"  # 이름, 0~1 회색조, 16진수
)
```

```python
# marker
fig, ax = plt.subplots()
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    ax.plot(
        np.random.rand(5), np.random.rand(5),
        marker=marker,
        label="marker={}".format(marker)
    )
ax.legend()
ax.set_xlim(0, 1.5)
```

```python
# 축 경계 조정
x = np.linspace(0, 10, 1000)

fig, ax = plt.subplots()
ax.plot(x, np.sin(x))
ax.set_xlim(-1, 11)
ax.set_ylim(-1.5, 1.5)

ax.set_aspect("equal")
```

```python
# 레이블 붙이기
x = np.arange(10)

fig, ax = plt.subplots()
ax.plot(x, x, label="y=x")
ax.plot(x, x**2, label="y=x^2")

ax.set_title("function")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend(loc="upper center", shadow=True)
```

```python
crime = pd.read_csv("../data/crime.csv")
```

```python
crime.head(5)
```

```python
gender_table = crime.pivot_table(index="젠더", columns="유형", values="명", aggfunc=np.sum)
```

```python
gender_table
```

```python
fig, ax = plt.subplots()
ax.set_title("2012~2016년 남여 피해자 통계")
ax.bar(["남자", "여자"], [gender_table.loc['남자', '피해자'], gender_table.loc['여자', '피해자']])
```

```python
fig, ax = plt.subplots()
ax.set_title("2012~2016년 남여 피해자 통계")
sns.barplot(
    ["남자", "여자"], [gender_table.loc['남자', '피해자'], gender_table.loc['여자', '피해자']],
    palette='deep'
)
```

```python
# 가해자와 피해자별 유형
```

```python
fig, ax = plt.subplots(1, 2, sharey=True)
ax[0].set_title("가해자")
sns.barplot(
    ["남자", "여자"],
    [gender_table.loc["남자", "가해자"], gender_table.loc["여자", "가해자"]],
    ax=ax[0], palette='deep'
)

ax[1].set_title("피해자")
sns.barplot(
    ["남자", "여자"],
    [gender_table.loc["남자", "피해자"], gender_table.loc["여자", "피해자"]],
    ax=ax[1], palette='deep'
)
```

```python
category_table = crime.pivot_table(
    index="젠더", columns=["대분류", "유형"], values='명', aggfunc=np.sum
)
```

```python
category_table
```

```python
data1 = category_table["강력범죄"]
data2 = category_table["지능범죄"]
data3 = category_table["폭력범죄"]
data4 = category_table["풍속범죄"]
```

```python
fig, ax = plt.subplots(1, 2, sharey=True)
index = ["남자", "여자"]

ax[0].set_title("가해자")
ax[0].bar(
    index, [data1.loc["남자", "가해자"], data1.loc["여자", "가해자"]],
    color='r', label="강력범죄"
)
bottom = np.array([data1.loc["남자", "가해자"], data1.loc["여자", "가해자"]])
ax[0].bar(
    index, [data2.loc["남자", "가해자"], data2.loc["여자", "가해자"]],
    color='g', label="지능범죄", bottom=bottom
)
bottom += np.array([data2.loc["남자", "가해자"], data2.loc["여자", "가해자"]])
ax[0].bar(
    index, [data3.loc["남자", "가해자"], data3.loc["여자", "가해자"]],
    color='lightblue', label="폭력범죄", bottom=bottom
)
bottom += np.array([data3.loc["남자", "가해자"], data3.loc["여자", "가해자"]])
ax[0].bar(
    index, [data4.loc["남자", "가해자"], data4.loc["여자", "가해자"]],
    color='b', label="풍속범죄", bottom=bottom
)

ax[1].set_title("피해자")
ax[1].bar(
    index, [data1.loc["남자", "피해자"], data1.loc["여자", "피해자"]],
    color='r', label="강력범죄"
)
bottom = np.array([data1.loc["남자", "피해자"], data1.loc["여자", "피해자"]])
ax[1].bar(
    index, [data2.loc["남자", "피해자"], data2.loc["여자", "피해자"]],
    color='g', label="지능범죄", bottom=bottom
)
bottom += np.array([data2.loc["남자", "피해자"], data2.loc["여자", "피해자"]])
ax[1].bar(
    index, [data3.loc["남자", "피해자"], data3.loc["여자", "피해자"]],
    color='lightblue', label="폭력범죄", bottom=bottom
)
bottom += np.array([data3.loc["남자", "피해자"], data3.loc["여자", "피해자"]])
ax[1].bar(
    index, [data4.loc["남자", "피해자"], data4.loc["여자", "피해자"]],
    color='b', label="풍속범죄", bottom=bottom
)
fig.legend(bbox_to_anchor=(1.1, 0.87), loc='upper right')
```

```python
category = crime.pivot_table(index="젠더", columns=["유형", "대분류"], values="명", aggfunc=np.sum)
```

```python
fig, ax = plt.subplots(1, 2, sharey=True)
category["가해자"].drop("불상").plot(kind='bar', stacked=True, ax=ax[0])
category["피해자"].drop("불상").plot(kind='bar', stacked=True, ax=ax[1])
```

```python
category = crime.groupby("대분류")
```

```python
violent_crime = category.get_group("강력범죄")
```

```python
cat2 = violent_crime.pivot_table(
    index="젠더", columns=["유형", "소분류"], values="명", aggfunc=np.sum
).drop("불상")
```

```python
fig, ax = plt.subplots(1, 2, sharey=True)
fig.suptitle("2012~2016 강력 범죄 가해자 및 피해자")
cat2['가해자'].plot(kind='bar', stacked=True, ax=ax[0], legend=False, rot=0)
cat2['피해자'].plot(kind='bar', stacked=True, ax=ax[1], legend=False, rot=0)
ax[1].legend(loc='upper right', bbox_to_anchor=(2, 1.04))
```

```python
years = violent_crime.pivot_table(
    index='연도', columns=["젠더", "유형"], values="명", aggfunc=np.sum
)[["남자", "여자"]]
```
