```python
shutil.rmtree("./univs")
os.makedirs("./univs")

city_groups = df.groupby("시도")

for city_name in df["시도"].unique():
    print(city_name)
    os.makedirs("./univs/{city_name}/".format(city_name=city_name))
    city_df = city_groups.get_group(city_name)
    
    univ_groups = city_df.groupby("학교명")
    for univ_name in city_df["학교명"].unique():
        print((city_name, univ_name))
        univ_df = univ_groups.get_group(univ_name)
        univ_df.to_excel("./univs/{city_name}/{univ_name}.xlsx".format(
            city_name=city_name,
            univ_name=univ_name
        ))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fire = df[(df['Type 1'] == 'Fire') | (df['Type 2'] == 'Fire')]
```

```python
water = df[(df['Type 1'] == 'Water') | (df['Type 2'] == 'Water')]
```

```python
print((len(fire), len(water)))
```

```python
fig, ax = plt.subplots()
ax.scatter(fire['Attack'], fire['Defense'], color='r', label='Fire', marker='*')
ax.scatter(water['Attack'], water['Defense'], color='b', label='Water')
ax.set_xlabel("Attack")
ax.set_ylabel("Defense")
ax.legend(loc="upper right")
```
