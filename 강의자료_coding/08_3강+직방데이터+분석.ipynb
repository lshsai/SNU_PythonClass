```python
import requests
import pandas as pd
```

```python
response = requests.get(url_a)
```

```python
response
```

```python
zigbang_df_A = pd.DataFrame(columns=["address", "deposit", "rent"])
```

```python
zigbang_df_A
```

```python
data = response.json()
```

```python
data
```

```python
for item in data["items"]:
    detail_info = item['item']
    
    deposit = detail_info.get("deposit")
    rent = detail_info.get("rent")
    address = detail_info.get("agent_address1")
    
    zigbang_df_A.loc[len(zigbang_df_A)] = [address, deposit, rent]
```

```python
zigbang_df_A
```

```python
res = requests.get(url_b)
```

```python
res
```

```python
zigbang_data = [
    item.get("item")
    for item
    in res.json().get("items")
]
```

```python
zigbang_df_B = pd.DataFrame(zigbang_data)
```

```python
zigbang_df_B = zigbang_df_B[["agent_address1", "deposit", "rent"]]
```

```python
zigbang_df_B = zigbang_df_B.rename(columns={"agent_address1": "address"})
```
