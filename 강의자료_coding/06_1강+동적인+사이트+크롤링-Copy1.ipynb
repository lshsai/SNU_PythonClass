```python
url = "https://watcha.net/home/news.json?page=1&per=100"
```

```python
res = requests.get(url)
```

```python
data = json.loads(res.text)
```

```python
data.get("news")[0].get("title")
```

```python
len(data.get("news"))
```

```python
file = open("watcha.csv", 'w')
file.write("title,url,content\n")

for news in data.get("news"):
    title = news.get("title")
    detail_url = news.get("detail_url")
    content = news.get("content")
    
    file.write("{},{},{}\n".format(
        title, detail_url, content
    ))
    print(title, content, detail_url)

file.close()
```

```python
import pandas as pd
items = []

for news in data.get("news"):
    title = news.get("title")
    detail_url = news.get("detail_url")
    content = news.get("content")
    
    temp_dict = {
        "title": title,
        "detail_url": detail_url,
        "content": content
    }
    items.append(temp_dict)
```

```python
watcha_df = pd.DataFrame(items)
```

```python
watcha_df.to_csv("watcha.csv")
```

```python
watcha_df.to_excel("watcha.xlsx")
```
